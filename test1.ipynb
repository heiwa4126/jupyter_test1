```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-math.pi, math.pi, 256)
plt.figure(figsize=(12, 6))
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```
